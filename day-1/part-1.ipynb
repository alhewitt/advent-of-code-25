```python
with open("input.txt") as f:
    rotations = f.readlines()
```

```python
def convert_rotation(rotation):
    direction = rotation[0]
    number = int(rotation[1:-1])
    # Can ignore the factor of 100 as it would be a whole rotation
    number = number % 100
    if direction == "L":
        number = -1*number
    return number
```

```python
dial = 50
code = 0
for rotation in rotations:
    dial += convert_rotation(rotation)
    if dial % 100 == 0:
        code += 1
print(code)
```
