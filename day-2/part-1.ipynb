```python
with open("input.txt") as f:
    range_list = f.readlines()
```

```python
import numpy as np

def get_ranges(range_list):
    return range_list[0].split(',')

def get_values_in_range(range):
    range = range.rstrip('\n').split('-') # Remove \n if applicable and then split
    start = int(range[0])
    end = int(range[1])
    return np.arange(start, end + 1)

def get_invalid_ids(values):
    invalid_ids = []
    for value in values:
        val = str(value)
        part_a, part_b = val[:len(val)//2], val[len(val)//2:]
        if part_a == part_b:
            invalid_ids.append(value)
    return invalid_ids

```

```python
ranges = get_ranges(range_list)
invalid_sum = 0

for range in ranges:
    values = get_values_in_range(range)
    invalid_ids = get_invalid_ids(values)
    invalid_sum += sum(invalid_ids)

print(invalid_sum)
```
