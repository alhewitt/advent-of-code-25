```python
with open("input.txt") as f:
    range_list = f.readlines()
```

```python
import numpy as np

def get_ranges(range_list):
    return range_list[0].split(',')

def get_values_in_range(rng):
    rng = rng.rstrip('\n').split('-') # Remove \n if applicable and then split
    start = int(rng[0])
    end = int(rng[1])
    return np.arange(start, end + 1)

def is_repeated_any(value: int) -> bool:
    val = str(value)
    length = len(val)
    for i in range(1, length // 2 + 1):
        if length % i != 0: # Skip non-factors
            continue
        n = length // i
        if n < 2: # Ensure there is some repetition
            continue
        part = val[:i]
        if part * n == val:
            return True
    return False

def get_invalid_ids(values):
    invalid_ids = []
    for value in values:
        if is_repeated_any(value):
            invalid_ids.append(value)
    return invalid_ids
```

```python
ranges = get_ranges(range_list)
invalid_sum = 0

for rng in ranges:
    values = get_values_in_range(rng)
    invalid_ids = get_invalid_ids(values)
    invalid_sum += sum(invalid_ids)

print(invalid_sum)
```
