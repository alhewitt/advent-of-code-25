```python
with open("input.txt") as f:
    rotations = f.readlines()
```

```python
def convert_rotation(rotation):
    direction = rotation[0]
    number = int(rotation[1:-1])
    if direction == "L":
        number = -1*number
    return number

def sign_change(a, b):
    return (a ^ b) < 0

def check_rotation(dial, rotation):
    step = 1 if rotation > 0 else -1
    zeros = 0
    position = dial
    for _ in range(abs(rotation)):
        position = (position + step) % 100
        if position == 0:
            zeros += 1
    return position, zeros
```

```python
dial = 50
code = 0
for rotation in rotations:
    raw_rotation = convert_rotation(rotation)
    dial, zeros = check_rotation(dial, raw_rotation)
    code += zeros
print(code)
```
